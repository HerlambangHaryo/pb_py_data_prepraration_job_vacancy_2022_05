```python

# set query 1
select = ' domain '
fromdb = 'jobstreet_raws_v2' 
where = '  group by domain '
orderby = '  order by domain ASC '
 
# query 1
mycursor = mydb.cursor()
mycursor.execute("SELECT " + select + " FROM `" + fromdb + "` " + where + orderby )
myresult = mycursor.fetchall()

print('We start and please wait ...')
# loop query 1
for x in myresult:   
    word = x[0]   
    
#     print(word)   
    
    queryx = "INSERT INTO domains (nama) VALUES ('"+word+"' )"
    print(queryx)
    mycursor.execute(queryx)
    mydb.commit()   
    
    
print('done : ' + str(len(myresult)) + ' row(s)')
```
