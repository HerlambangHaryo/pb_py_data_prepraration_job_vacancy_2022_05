```python
# Connect to DB
host="localhost"
user="root"

# host="192.168.100.3"
# user="rootx"

import mysql.connector
mydb = mysql.connector.connect(
    host=host,
    user=user,
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)
mycursor = mydb.cursor()
print('connect')
```

```python
# query 1 
print('We start and please wait ...')

word_pre = 'Acounting'
word_aft = 'Accounting'
# query 1
entity = 'title'
select = 'id, ' + entity
fromdb = 'jobstreet_raws_v2'
where = "where "+entity+" like '%"+word_pre+"%' "
orderby = " Order By "+entity+" ASC"

mycursor = mydb.cursor()
mycursor.execute("SELECT " + select + " FROM `" + fromdb + "` " + where + orderby )
myresult = mycursor.fetchall()

# loop query 1
for x in myresult:    
    idx = str(x[0])   
    word = x[1]  
    word_fnl = word.replace(word_pre, word_aft)
    print(word)   
    print(word_fnl)   
    
    queryx = "UPDATE "+fromdb+" SET "+entity+" = '" + word_fnl + "'   " 
    wherex = "  where id = '" + idx + "'"  
    
    mycursor.execute(queryx + wherex)
    mydb.commit()   

print('done : ' + str(len(myresult)) + ' row(s)')
print("")   
print("############################################################################################")
#########################################################################################################
# query 2 
print('We start and please wait ...')
fromdb = 'titles_raws' 

entity = 'nama'
select = 'id, ' + entity
where = "where "+entity+" like '%"+word_pre+"%' "
orderby = " Order By "+entity+" ASC"
mycursor = mydb.cursor()
mycursor.execute("SELECT " + select + " FROM `" + fromdb + "` " + where + orderby )
myresult = mycursor.fetchall()

# loop query 2
for x in myresult:   
    idx = str(x[0])   
    word = x[1]  
    word_fnl = word.replace(word_pre, word_aft)
    print(word)   
    print(word_fnl)   
    
    queryx = "UPDATE "+fromdb+" SET "+entity+" = '" + word_fnl + "'   " 
    wherex = "  where id = '" + idx + "'"  
    
    mycursor.execute(queryx + wherex)
    mydb.commit()   
    
print('done : ' + str(len(myresult)) + ' row(s)')
```
