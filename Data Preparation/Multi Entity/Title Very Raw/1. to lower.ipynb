```python
# Connect to DB

host="localhost"
user="root"

# host="192.168.100.3"
# user="rootx"

import mysql.connector
mydb = mysql.connector.connect(
    host=host,
    user=user,
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)
mycursor = mydb.cursor()
print('connect')

```

```python

# set query 1
select = ' id, nama '
fromdb = 'title_very_raws' 
where = '  '
orderby = '  order by nama ASC '
 
# query 1
mycursor = mydb.cursor()
mycursor.execute("SELECT " + select + " FROM `" + fromdb + "` " + where + orderby )
myresult = mycursor.fetchall()

print('We start and please wait ...')
# loop query 1
for x in myresult:  
    idx = str(x[0]) 
    word = x[1].lower()   
    
    print(word)   
    
    queryx = "UPDATE title_very_raws  SET nama = '"+word+"' where id = '"+idx+"'    " 
    mycursor.execute(queryx)
    mydb.commit()   
    
    
print('done : ' + str(len(myresult)) + ' row(s)')
```
