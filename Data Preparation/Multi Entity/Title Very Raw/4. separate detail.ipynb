```python
# Connect to DB

host="localhost"
user="root"

# host="192.168.100.3"
# user="rootx"

import mysql.connector
mydb = mysql.connector.connect(
    host=host,
    user=user,
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)
mycursor = mydb.cursor()
print('connect')

```

```python

# set query 1
select = ' id, nama '
fromdb = 'title_very_raws' 
where = '  '
orderby = '    '
 
# query 1
mycursor = mydb.cursor()
mycursor.execute("SELECT " + select + " FROM `" + fromdb + "` " + where + orderby )
myresult = mycursor.fetchall()

print('We start and please wait ...')
# loop query 1
for x in myresult:  
    idx = str(x[0]) 
    word = x[1]
    
#     print(word)   
    
    check_word = '('
    if(check_word in word):
        word_0 = word.split(check_word)[0]
        word_1 = word.split(check_word)[1]
        
        word_0 = word_0
        
        if(word_0 != '' and word_1 != ''): 
            queryx = "UPDATE title_very_raws  SET nama  = '"+word_0+"', nama_detail  = '"+word_1+"' where id = '"+idx+"'    " 
            print(queryx)
            mycursor.execute(queryx)
            mydb.commit()   
    
    
    
    
    
print('done : ' + str(len(myresult)) + ' row(s)')
```
