```python
import mysql.connector

mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)

select = ' * '
frdb = 'wilayah_2022'
where = ' WHERE LENGTH(kode) = 2 '  
 
mycursor = mydb.cursor()

mycursor.execute("SELECT " + select + " FROM `" + frdb + "` " + where )

myresult = mycursor.fetchall()

for x in myresult:   
    words = x[1] + " " + x[2]   
    sql = "INSERT INTO provinsis (kode, nama, created_at) VALUES ('"+x[1]+"', '"+x[2].title()+"', now())"
    mycursor.execute(sql)
    mydb.commit()  
    print(words)
    

print('Done')
     
```

```python
import mysql.connector

mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)

sql = ' select * From kotas  '  
 
mycursor = mydb.cursor()
mycursor.execute(sql)
myresult = mycursor.fetchall()

for x in myresult:   
    words = x[1] + " " + x[2]   
    print(words)
    
    sql2 = ' select * From wilayah_2022 where kode like "' + str(x[1]) + '.%" and LENGTH(kode) = 8 '  

    mycursor2 = mydb.cursor()
    mycursor2.execute(sql2)
    myresult2 = mycursor2.fetchall()
    
    for y in myresult2:   
        kecamatan_name = y[2].replace("'", "\\'")
        sql3 = "INSERT INTO kecamatans (kota_id, kode, nama) VALUES ('" + str(x[0])+ "', '"+ str(y[1]) + "', '" + str(kecamatan_name.title()) + "' ) "   
        mycursor3 = mydb.cursor()
        mycursor3.execute(sql3)
        mydb.commit()  
        words = str(x[0]) + " " + str(y[1]) + " " + str(y[2])    
        print(words)
    

print('Done')
     
```

```python
import mysql.connector

mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)

sql = ' select * From kecamatans  '  
 
mycursor = mydb.cursor()
mycursor.execute(sql)
myresult = mycursor.fetchall()

for x in myresult:   
    words = "lv-x " + x[1] + " " + x[2]   
    print(words)
    
    sql2 = ' select * From wilayah_2022 where kode like "' + str(x[2]) + '.%" and length(kode) >= 10 '  

    mycursor2 = mydb.cursor()
    mycursor2.execute(sql2)
    myresult2 = mycursor2.fetchall()
    
    for y in myresult2:   
        kelurahan_name = y[2].replace("'", "\\'")
        sql3 = "INSERT INTO kelurahans (kecamatan_id, kode, nama) VALUES ('" + str(x[0])+ "', '"+ str(y[1]) + "', '" + str(kelurahan_name.title()) + "' ) "   
        mycursor3 = mydb.cursor()
        mycursor3.execute(sql3)
        mydb.commit()  
        words = "lv-y " +str(x[0]) + " " + str(y[1]) + " " + kelurahan_name   
        print(words)
    

print('Done')
     
```
