```python
# Connect DB

import mysql.connector
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)
mycursor = mydb.cursor()
print('connect')
```

```python
select = ' id, title '
fromdb = 'jobstreet_raws_rc' 
where = ' '
group = ' ' 
orderby = ' order by id asc ' 
 
mycursor = mydb.cursor()
mycursor.execute("SELECT " + select + " FROM `" + fromdb + "` " + where + group + orderby  )
myresult = mycursor.fetchall()

for x in myresult:    
    ids = str(x[0]) 
    title_rep = x[1].replace(r"\u002F", " / ")
    title = title_rep.strip().title()   
    print(ids + " " + title)
    sql = "INSERT INTO job_vacancies (title) VALUES ( '" + title + "')"
    print(sql)
    mycursor.execute(sql)
    mydb.commit()   
```
