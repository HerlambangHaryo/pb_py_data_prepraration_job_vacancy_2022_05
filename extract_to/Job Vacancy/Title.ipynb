```python
# Connect to DB

import mysql.connector
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="pr_scraping_job_vacancy_indonesia_2022_04"
)

mycursor = mydb.cursor()

print('connect')
```

```python
# update job_vacancies 
entity = 'title'

# set query 1
select = ' id, ' + entity
fromdb = 'jobstreet_raws_v2'
where = ' '
group = ' ' 
 
# query 1
mycursor = mydb.cursor()
mycursor.execute("SELECT " + select + " FROM `" + fromdb + "` " + where + group  )
myresult = mycursor.fetchall()
   
# loop query 1
for x in myresult:  
    idx = str(x[0])
    word_temp = x[1] 
    word = word_temp.replace("'", "\\'")
    
    # update extract to
    query2 = "UPDATE job_vacancies SET "+entity+" = '" + word + "' " 
    where2 = " where raw_id = " + idx + "  " 
    print(query2 + where2)
    mycursor.execute(query2 + where2 )
    mydb.commit()  

    # update extracted
    queryx = "UPDATE jobstreet_raws_rc SET "+entity+" = '--extracted--'   " 
    wherex = "  where id = '" + idx + "'"  
    print(queryx + wherex)
    mycursor.execute(queryx + wherex)
    mydb.commit()  
     
         
print('done : ' + str(len(myresult)) + ' row(s)')
```
